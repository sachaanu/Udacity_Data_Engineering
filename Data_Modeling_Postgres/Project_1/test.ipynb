```python
%sql SELECT RIGHT(songplay.location, 2) AS state, COUNT(songplay.session_id) AS freq FROM songplay JOIN users ON users.user_id = songplay.user_id GROUP BY 1 ORDER BY 2 DESC
```

```python
%sql SELECT users.first_name, users.last_name, tmp.state, tmp.freq AS frequency, tmp.rw AS Rank_in_state \
FROM users, \
(SELECT \
        DISTINCT(users.user_id), \
        RIGHT(songplay.location, 2) AS state, \
        COUNT(songplay.session_id) AS freq, ROW_NUMBER() OVER(PARTITION BY RIGHT(songplay.location, 2) ORDER BY COUNT(songplay.session_id) DESC) as rw \
        FROM songplay JOIN users ON users.user_id = songplay.user_id \
        GROUP BY 2,1 \
        ORDER BY 3 DESC) tmp \
WHERE users.user_id = tmp.user_id \
AND tmp.rw <= 2 \
ORDER BY tmp.state ASC, tmp`.freq DESC;
```
